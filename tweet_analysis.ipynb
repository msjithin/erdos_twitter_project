```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
apple_df = pd.read_csv("data\Tweets_Raw\df_Apple.csv")
dollar_df = pd.read_csv("data\Tweets_Raw\df_Facebook.csv")
```

```python
apple_df.sample()
```

```python
dollar_df.sample()
```

```python
#apple_df = apple_df.set_index('created_at')
apple_df['Date']=pd.to_datetime(apple_df['created_at'])
dollar_df['Date']=pd.to_datetime(dollar_df['created_at'])
```

```python
apple_df.set_index('Date',inplace=True)
dollar_df.set_index('Date',inplace=True)
```

```python
print(apple_df.columns)
```

```python
apple_df['public_metrics_like_count'].plot(label="likes", c="r")
apple_df['public_metrics_retweet_count'].plot(label="retweets", c="b")
apple_df['public_metrics_reply_count'].plot(label="reply", c="g")

plt.legend(loc='best')
```
